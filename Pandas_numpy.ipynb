```python
#concat
df1= pd.DataFrame(
    {'A':['A0','A1','A2','A3'],
    'B':['B0','B1','B2','B3'],
    'C':['C0','C1','C2','C3'],
    'D':['D0','D1','D2','D3']
})
```

```python
#concat
df2= pd.DataFrame(
    {'A':['A4','A5','A6','A7'],
    'B':['B4','B5','B6','B7'],
    'C':['C4','C5','C6','C7'],
    'D':['D4','D5','D6','D7']
})
```

```python
df1
```

```python
df2
```

```python
result = pd.concat([df1, df2])
```

```python
result
```

```python
#loc vs iloc
#loc gets rows with particular labels
#iloc gets rows with integer locations
```

```python
s = pd.Series(list("abcdef"),index=[49,48,47,0,1,2])
s
```

```python
s.loc[0] #value at index label 0
```

```python
s.iloc[0] #value at index location 0
```

```python
s.loc[0:1] #rows at index lables between 0 and 1 (inclusive)
```

```python
s.iloc[0:1] #rows at location 0 and 1(exclusive)
```
