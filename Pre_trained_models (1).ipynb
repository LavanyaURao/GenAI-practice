```python
from keras.applications import VGG16
from tensorflow.keras.preprocessing import image
from keras.applications.vgg16 import preprocess_input, decode_predictions
import numpy as np
```

```python
#load VGG16 pretrained model and specify the input shape
model = VGG16(weights='imagenet',include_top=True, input_shape=(224,224,3))
```

```python
img_path= '/content/hi.png'
img = image.load_img(img_path, target_size=(224,224,3))
```

```python
x = image.img_to_array(img)
x = np.expand_dims(x, axis=0)
x = preprocess_input(x)
```

```python
#use the model to predict the class

predict = model.predict(x)
```

```python
predict
```

```python
#get the top 5 pred with class names

decode_pred = decode_predictions(predict, top=10)[0]
```

```python
for pred in decode_pred:
  print(pred[1],":",pred[2])
```

---

chks each square n tells if its the image we r trying to identify. to know whc square is imp.

Depth wise convolution- first we use this. puts a circle n highlights all the imp areas. Finds important features in the image channel by channel.


pointwise convolution- Uses a 1×1 filter to combine the features obtained from depthwise convolution.






Imagine an X-ray image:

1. **Depthwise Convolution**

Looks at each feature separately and says:

"This edge is important."
"This texture is important."

2. **Pointwise Convolution**

Combines all those observations and says:

"These features together indicate pneumonia."
