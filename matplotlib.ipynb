```python
churn_df = pd.read_csv('Churn_Modelling.csv')
```

```python
churn_df.head(3)
```

```python
churn_df.Geography.value_counts()
```

```python
#churn_df.Geography.value_counts() is what we use
churn_df.Geography.value_counts().plot(kind='barh')
```

```python
#for exited column
churn_df['Exited'].value_counts().plot(kind='bar')
```

```python
churn_df['Exited'].value_counts()
```

```python
#to find total percentage
churn_df['Exited'].value_counts()/len(churn_df)*100
```

```python
plt.bar(churn_df['Gender'], churn_df['Age'].mean(),color="cyan" , edgecolor="orange")
plt.title('Gender vs Age', fontdict = {"fontsize": 20, "color": "pink"})
```

```python
churn_df.Tenure
```

```python
plt.scatter(churn_df['Age'], churn_df['Tenure'])
plt.title('Scatter plot')
plt.xlabel('This is x axis')
plt.ylabel('This is y axis')
plt.show()
```

```python
#histogram
plt.hist(churn_df['Tenure'], bins=30)
plt.xlabel('Probability')
plt.ylabel('Tenure')
```

```python
labels_ex = 'JavaScript', 'Java', 'Python', 'R'
sizes = [15, 30, 45, 10]
explode_labels= (0, 0.2,0,0)
fig1, ax1 = plt.subplots()

ax1.pie(sizes, explode = explode_labels, labels = labels_ex, shadow= True, startangle=90)
ax1.axis('equal')
plt.show()
```
