## GRU Gates вҖ” Simple Understanding

A **GRU (Gated Recurrent Unit)** has **2 main gates**:

1. **Reset Gate**
2. **Update Gate**

These gates help the model remember important information and forget unnecessary information.

---

**1. Reset Gate**

Controls **how much past information should be forgotten**.

means-

> вҖңShould I ignore the old memory?вҖқ

* If reset gate value is **0** вҶ’ forget old information
* If reset gate value is **1** вҶ’ keep old information

### Example:

Suppose sentence is:

> "I grew up in France... I speak fluent French."

The model should remember **France**.

But in:

> "I grew up in France... Now I live in India."

Some old information may become less important.

The **reset gate** decides this.

---

**2. Update Gate**

Controls **how much new information should be added**.

means-

> вҖңShould I keep old memory or update with new memory?вҖқ

* Value close to **1** вҶ’ keep old memory
* Value close to **0** вҶ’ replace with new memory

### Example:

In sentence:

> "The movie was boring but the ending was amazing."

The word **amazing** changes the meaning.

The update gate helps the GRU update its understanding.

---
**Reset Gate**

Determines how much past information to forget.

**Update Gate**

Determines how much new information to keep and how much old information to retain.

---
GRU-
* Simpler than LSTM
* Faster training
* Uses only 2 gates
* Works well for sequence data and NLP tasks


```python
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, GRU, Dense

text = """
The Royal Challengers Bengaluru, also known as RCB (formerly Royal Challengers Bangalore),
are a professional Twenty20 cricket team based in Bengaluru, Karnataka, that competes in the
Indian Premier League (IPL).

They won their first IPL title in 2025.

The team finished as runners-up on three occasions:
- 2009
- 2011
- 2016

RCB has also qualified for the playoffs in ten of the eighteen seasons.

As of 2026:
- Captain: Rajat Patidar
- Coach: Andy Flower

The franchise has also competed in the Champions League Twenty20, where they finished as runners-up in 2011.

As of 2024, RCB's brand value was estimated at $117 million, making it one of the most valuable IPL franchises.

Founded in 2008 by United Spirits, the franchise has been owned by:
- Aditya Birla Group
- The Times Group
- Blackstone (since 2026)

The team was reportedly acquired for $1.78 billion, marking the highest valuation for an IPL franchise.
"""

tokenizer = Tokenizer()
tokenizer.fit_on_texts([text])

# Total vocabulary size
vocab_size = len(tokenizer.word_index) + 1



input_sequences = []

for line in text.split('\n'):

    token_list = tokenizer.texts_to_sequences([line])[0]

    for i in range(1, len(token_list)):
        n_gram_sequence = token_list[:i + 1]
        input_sequences.append(n_gram_sequence)


max_len = max([len(x) for x in input_sequences])

padded_input_sequences = pad_sequences(
    input_sequences,
    maxlen=max_len,
    padding='pre'
)


X = padded_input_sequences[:, :-1]
y = padded_input_sequences[:, -1]

#models

model = Sequential()

# Embedding Layer
model.add(
    Embedding(
        input_dim=vocab_size,
        output_dim=100,
        input_shape=(X.shape[1],)
    )
)

# LSTM Layer
model.add(GRU(150))

# Output Layer
model.add(Dense(vocab_size, activation='softmax'))


model.compile(
    loss='sparse_categorical_crossentropy',
    optimizer='adam',
    metrics=['accuracy']
)


```

```python
model.summary()
```

```python
model.fit(X,y, epochs =100)
```

```python
text2= "Rcb"

#tokenization
token_text= tokenizer.texts_to_sequences([text2])[0]

#padding
padded_text= pad_sequences([token_text], maxlen=max_len,padding='pre' )

#predict
model.predict(padded_text)
```

```python
import numpy as np
np.argmax(model.predict(padded_text))
```
