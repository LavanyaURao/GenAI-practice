**PoS Tagging- parts of speech**

Ram(noun) is going(verb)

task of labelling each word in a sentance with its appropriate part of speech.

Applications-

1. name entity recognition

2. Q and A system

3. Chatbots

4. Word sense disambiguation
image.png

spacy- entitys exact tags identifiction


```python
#NER
import spacy
```

```python
nlp = spacy.load('en_core_web_sm')
```

```python
nlp.pipe_names
```

```python
text = nlp("Tesla is going to acquire twitter for $45 billion")

for ent in text.ents:
  print(ent.text, " ? ", ent.label_, " | ", spacy.explain(ent.label_))
```

```python
text = nlp("Tesla is going to acquire twitter for forty five billion dollars")

for ent in text.ents:
  print(ent.text, " ? ", ent.label_, " | ", spacy.explain(ent.label_))
```

```python
from spacy import displacy
displacy.render(text, style="ent")
```

```python
text2= nlp("Lavanya will go to india")
displacy.render(text2, style="ent")
```

Setting up own entities

```python
doc = nlp("Tesla is going to acquire twitter for $45 billion")

for ent in text.ents:
  print(ent.text, " ? ", ent.label_, " | ", spacy.explain(ent.label_))
```

```python
s = doc[3:6]
s
```

```python
from spacy.tokens import Span

s1 = Span(doc, 0, 2, label="ORG")
s2 = Span(doc, 8, 9, label="ORG")

doc.set_ents([s1, s2], default="unmodified")

for ent in doc.ents:
  print(ent.text, " ? ", ent.label_, " | ", spacy.explain(ent.label_))
```
